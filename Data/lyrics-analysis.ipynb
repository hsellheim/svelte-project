I want to analyze whether darker or lighter colors are more prominent, in general and per album, to determin whether there are any patterns. I am categorizing them in this way: 
<ul>
<li>dark colors: black, brown, grey, purple, green, blue</li>
<li>light colors: red, yellow, pink, white, gold, silver, orange</li>
</ul>

```python
dark_color_counts = (sum(df_albums['black']) + sum(df_albums['brown']) + sum(df_albums['grey']) + sum(df_albums['purple']) + sum(df_albums['green']) + sum(df_albums['blue']))
```

```python
dark_color_counts
```

```python
light_color_counts = (sum(df_albums['red']) + sum(df_albums['yellow']) + sum(df_albums['pink']) + sum(df_albums['white']) + sum(df_albums['gold']) + sum(df_albums['silver']) + sum(df_albums['orange']))
```

```python
light_color_counts
```

```python
df_ldc = pd.DataFrame(columns = ['light', 'dark'])
```

```python
df_ldc['light'] = [light_color_counts]
```

```python
df_ldc['dark'] = [dark_color_counts]
```

```python
df_ldc
```

```python
df_ldc.to_csv('light-dark.csv', index=False)
```
